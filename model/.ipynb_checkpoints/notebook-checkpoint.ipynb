Step 4

Business setting :
We are a flower market and we need to sort the flowers that we get from the suppliers.

So between : Iris-setosa (0) or others (1)

Step 5

Iris setosa is the product we sell at the highest price so we need to identify it.
A false positive would mean that we are classifying another iris as a setosa (0).
A false negative would mean not recognising the iris setosa when we should have. 

The consequences of a false positive would be damaging our relationship with the client because the product is not what we announced.
The consequences of a false negative would mean a loss of revenue (setose is our best product and the selling price of is higher than the other iris flowers we sell)

Damaging relationship with the customer can make us lose business on a long term scale, so we want to weight FP two times:

FP = FN * 2

selling price:
setosa = 10
other iris = 8

```python
#Step 6

def calculate_cost(y_actual, y_pred):

    FP = count((y_actual == 0) & (y_pred == 1))
    FN = count((y_actual == 1) & (y_pred == 0))

    cost = (FP*5)*2 + FN*2

    return cost

#The code essentially penalizes false positives more heavily than false negatives
#when calculating the cost of a binary classification model.
```
